**Question 1 : What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?**
- In hypothesis testing, we begin by setting up two competing statements: the null hypothesis (H‚ÇÄ), which usually represents ‚Äúno effect‚Äù or ‚Äúno difference,‚Äù and the alternative hypothesis (H‚ÇÅ), which represents the effect or difference we are trying to detect. When we make decisions based on sample data, there is always a chance of drawing the wrong conclusion. These incorrect conclusions are known as Type I and Type II errors, and understanding them is essential for making reliable statistical decisions.

Type I Error (Œ± Error)

A Type I error occurs when we reject the null hypothesis even though it is actually true. In simple terms, the test claims that an effect or difference exists when in reality it does not. This is often called a false positive. The probability of committing a Type I error is denoted by Œ±, which is also known as the significance level.

For example, if Œ± = 0.05, we are allowing a 5% risk of incorrectly concluding that there is a significant effect. Type I errors are critical in fields where false alarms can lead to serious consequences, such as approving a harmful drug, detecting a disease that is not present, or making unnecessary business changes.

Type II Error (Œ≤ Error)

A Type II error occurs when we fail to reject the null hypothesis even when the alternative hypothesis is true. This means that the test fails to detect a real effect or difference. It is also called a false negative. The probability of making a Type II error is represented by Œ≤, and the complement (1 ‚Äì Œ≤) is known as the power of the test, which reflects the ability of the test to detect an actual effect.

Type II errors are problematic in situations where missing a true effect has negative outcomes. For example, failing to detect a defective product, not identifying a disease that a patient actually has, or overlooking a meaningful improvement in a business process.

Balancing Type I and Type II Errors

There is a trade-off between Œ± and Œ≤. Reducing the chance of a Type I error often increases the chance of a Type II error, unless the sample size is increased. Researchers must choose the level of Œ± based on the seriousness of the potential error. In situations where false positives are very costly, Œ± is set lower (such as 0.01). If missing a true effect is more serious, then increasing sample size or accepting a slightly higher Œ± can help improve test power.

Impact on Decision-Making

Understanding Type I and Type II errors is essential for making sound decisions based on statistical tests:

Policy decisions: Governments or institutions may set stricter Œ± levels if a false approval is highly risky.

Medical testing: Type I error means falsely diagnosing a disease; Type II means missing an actual disease. The acceptable error depends on the medical context.

Business analytics: A Type I error might trigger unnecessary marketing changes, while a Type II error might cause a company to miss valuable opportunities.

**Question 2:What is the P-value in hypothesis testing, and how should it be interpreted**
in the context of the null hypothesis?
- In hypothesis testing, the P-value plays a central role in determining whether the evidence provided by the sample data is strong enough to reject the null hypothesis (H‚ÇÄ). It provides a quantitative measure of the strength of evidence against the null hypothesis, helping researchers make objective statistical decisions rather than relying on subjective judgment.

Definition of the P-value

The P-value is defined as the probability of obtaining a test statistic (or result) that is as extreme as or more extreme than the one observed in the sample data, assuming that the null hypothesis is true.
In other words, it answers the question:

‚ÄúIf the null hypothesis were correct, how likely is it to observe sample results like the ones we got‚Äîor more extreme?‚Äù

Because it is a probability, the P-value always lies between 0 and 1.

Role of the P-value in Hypothesis Testing

When conducting hypothesis tests, researchers compare the P-value to a predetermined significance level (Œ±), often set at 0.05, 0.01, or 0.10. The comparison guides the decision to either reject or fail to reject the null hypothesis.

If P-value ‚â§ Œ±, there is enough evidence to reject H‚ÇÄ.

If P-value > Œ±, we fail to reject H‚ÇÄ.

This comparison provides a consistent method to decide whether the observed effect is statistically significant.

Interpreting the P-value

Small P-value (typically < 0.05)
A small P-value indicates that the observed sample result would be very unlikely if the null hypothesis were true.
This suggests strong evidence against H‚ÇÄ, and we reject the null hypothesis.
A small P-value does not prove that H‚ÇÄ is false but shows that the data contradict it.

Large P-value (typically > 0.05)
A large P-value suggests that the observed data are consistent with the null hypothesis.
This does not prove that H‚ÇÄ is true; it merely indicates that the sample does not provide enough evidence to reject it.
In other words, the data do not contradict the null hypothesis strongly enough.

Moderate P-value (around 0.05)
Values near the chosen significance level require careful interpretation. Researchers should consider context, sample size, practical significance, and study design.

Common Misconceptions About P-values

The P-value is often misunderstood. It is important to clarify what it does not represent:

It does not give the probability that the null hypothesis is true.

It does not measure the size or importance of an effect.

It is not the probability that the observed results are due to chance alone.

A large P-value is not evidence that the null hypothesis is definitely true.

The P-value strictly reflects the probability of observing the sample results under the assumption that the null hypothesis is true.

Contextual Interpretation

The interpretation of the P-value must always be tied to the null hypothesis being tested. For example:

If H‚ÇÄ states that the population mean is 50, a P-value of 0.03 means that if the true mean were indeed 50, the probability of getting a sample mean as extreme as the one observed is only 3%.

This low probability indicates that such data are unusual under H‚ÇÄ, making it reasonable to reject H‚ÇÄ and infer that the population mean is likely not 50.

Practical Importance

P-values are widely used across scientific research, medicine, economics, and business analytics because they give a standardized way to judge statistical evidence. However, they should be used along with:

Effect sizes

Confidence intervals

Study design considerations

Practical significance

to make well-rounded decisions.

Question 3:Explain the difference between a Z-test and a T-test, including when to use
each. **bold text**
- Hypothesis testing often involves comparing sample statistics with population parameters to determine whether an observed effect is statistically significant. Among the most widely used statistical tests for comparing means are the Z-test and the T-test. Although they share a similar purpose, they differ in their underlying assumptions, formulas, and applicability. Understanding these differences is crucial for selecting the appropriate test in real-world data analysis.

1. Conceptual Difference Between Z-test and T-test
Z-test

A Z-test is used when the sample size is large (typically n ‚â• 30) or when the population standard deviation (œÉ) is known. The Z-distribution is a standard normal distribution with mean 0 and standard deviation 1. Because of the Central Limit Theorem, the sampling distribution of the mean becomes approximately normal for large samples regardless of the population distribution.

T-test

A T-test, developed by William Gosset under the pseudonym ‚ÄúStudent,‚Äù is used when the population standard deviation is unknown and the sample size is small (n < 30). Instead of the normal distribution, it uses the t-distribution, which has thicker tails. This accounts for the extra uncertainty that arises from estimating the population standard deviation using the sample standard deviation (s).

2. Distributional Differences
Z-distribution

Symmetrical and exactly normal.

Has fixed shape regardless of sample size.

Less variability because the population standard deviation is assumed to be known.

T-distribution

Also symmetrical but has heavier tails.

The shape depends on the degrees of freedom (df = n ‚Äì 1).

As sample size increases, the t-distribution approaches the normal distribution.

More appropriate for small samples because it captures extra uncertainty.
4. When to Use Each Test
Use a Z-test when:

The population standard deviation (œÉ) is known.

The sample size is large (n ‚â• 30).

The population is normally distributed or the sample is large enough for the Central Limit Theorem to apply.

You are conducting tests such as:

Tests for population mean

Tests for population proportion

Large sample comparisons

Z-tests are commonly used in quality control, industrial processes, population surveys, and large-scale experiments where œÉ is known from historical data.

Use a T-test when:

The population standard deviation (œÉ) is unknown.

The sample size is small (n < 30).

The sample comes from a population that is approximately normally distributed.

You are conducting:

One-sample t-test

Independent two-sample t-test

Paired t-test

T-tests are widely used in controlled experiments, academic research, clinical trials, and smaller datasets where œÉ is not available.

5. Why the T-test Is Needed for Small Samples

When œÉ is estimated from the sample as s, the estimate has variability, and this uncertainty increases when the sample size is small. The t-distribution compensates for this uncertainty with its heavier tails. This means that for the same confidence level, t-tests have larger critical values than Z-tests. This prevents underestimating the probability of extreme values, ensuring more reliable conclusions.

6. Convergence of T-test to Z-test

As sample size increases, the difference between t-distribution and Z-distribution becomes negligible. For large samples (n ‚â• 30), the t-test and Z-test give nearly identical results because s becomes a good approximation of œÉ. Therefore, the choice of test matters most when the sample is small.

ANOVA does not identify which specific groups differ; it only tells us whether a significant difference exists somewhere. If the ANOVA shows significance, researchers usually follow up with post-hoc tests (e.g., Tukey HSD, Bonferroni correction) to identify the specific group differences.

2. Why ANOVA Is Needed for More Than Two Groups

If a researcher uses multiple t-tests to compare several group means, the probability of making a Type I error increases significantly. For example, comparing 4 groups using pairwise t-tests results in 6 comparisons, each with its own risk of error. This inflates the chance of finding a false significant result.

ANOVA solves this problem by providing a single overall test that compares all group means simultaneously while controlling the Type I error rate.

Thus, ANOVA is an extension of hypothesis testing that remains robust and reliable when dealing with multiple group comparisons.

4. Assumptions of ANOVA

To ensure the validity of ANOVA results, certain assumptions must be met. These assumptions relate to the structure, distribution, and variance of the data.

1. Independence of Observations

Each observation must be independent of the others. This assumption is typically satisfied through proper random sampling or random assignment in experiments.

2. Normality

The data within each group should be approximately normally distributed. ANOVA is robust to slight violations of normality, especially when sample sizes are large and similar across groups.

3. Homogeneity of Variances (Homoscedasticity)

The variance in each group should be approximately equal. This assumption ensures that the within-group variability is comparable across groups. Statistical tests like Levene‚Äôs test or Bartlett‚Äôs test are commonly used to check this assumption.

4. Measurement Scale

The dependent variable should be measured on an interval or ratio scale.

When these assumptions are reasonably met, ANOVA provides valid and reliable results. If assumptions are violated, alternative tests like Welch‚Äôs ANOVA or non-parametric Kruskal‚ÄìWallis tests may be used.

5. Types of ANOVA

ANOVA is not limited to one type. The most common forms include:

One-Way ANOVA

Used when comparing the means of three or more groups based on one independent variable (factor).
Example: Comparing mean test scores of students from four teaching methods.

Two-Way ANOVA

Used when analyzing the influence of two different independent variables on one dependent variable, including possible interactions.
Example: Impact of teaching method and gender on test scores.

Repeated Measures ANOVA

Used when the same subjects are measured multiple times under different conditions.
Example: Measuring heart rate of patients before, during, and after treatment.

6. Extension Beyond Two Groups

ANOVA extends hypothesis testing beyond two groups by using a single global test that:

Controls overall Type I error

Analyzes variation across multiple groups

Provides a structured framework for identifying overall differences

This means ANOVA allows researchers to test:

Three groups (e.g., low, medium, high dosage)

Four groups (e.g., four teaching methods)

Ten groups, or even more

without increasing the risk of incorrect conclusions.

If ANOVA indicates statistical significance, researchers proceed to post-hoc tests:

Tukey HSD

Scheff√© test

Bonferroni correction

Duncan test

These tests locate the specific pairs of group means that differ while keeping error rates controlled.




```python
#Question 6: Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset
# one_sample_z_test.py
import numpy as np
from math import sqrt
import scipy.stats as stats

def one_sample_z_test(sample, popmean, popstd, alpha=0.05, alternative='two-sided'):
    """
    Performs a one-sample Z-test.
    sample: array-like sample values
    popmean: population mean under H0
    popstd: known population standard deviation (sigma)
    alpha: significance level
    alternative: 'two-sided', 'less', or 'greater'
    Returns: dict with z_stat, p_value, reject_H0 (bool), ci (95% default)
    """
    x = np.asarray(sample)
    n = len(x)
    xbar = np.mean(x)
    se = popstd / sqrt(n)
    z_stat = (xbar - popmean) / se

    if alternative == 'two-sided':
        p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))
        crit = stats.norm.ppf(1 - alpha/2)
        ci_low = xbar - crit * se
        ci_high = xbar + crit * se
    elif alternative == 'less':
        p_value = stats.norm.cdf(z_stat)
        crit = stats.norm.ppf(alpha)
        ci_low = -np.inf
        ci_high = xbar + stats.norm.ppf(1 - alpha) * se
    elif alternative == 'greater':
        p_value = 1 - stats.norm.cdf(z_stat)
        crit = stats.norm.ppf(1 - alpha)
        ci_low = xbar - stats.norm.ppf(1 - alpha) * se
        ci_high = np.inf
    else:
        raise ValueError("alternative must be 'two-sided', 'less', or 'greater'")

    reject = p_value < alpha

    return {
        'n': n,
        'sample_mean': xbar,
        'z_stat': z_stat,
        'p_value': p_value,
        'reject_H0': reject,
        'conf_interval': (ci_low, ci_high)
    }

# Example usage:
if __name__ == "__main__":
    # Example dataset
    sample = [5.1, 4.9, 5.0, 5.4, 4.8, 5.2, 5.0, 5.3]
    popmean = 5.0            # H0: mu = 5.0
    popstd = 0.5             # assumed known population std dev
    res = one_sample_z_test(sample, popmean, popstd, alpha=0.05, alternative='two-sided')
    print("Results:", res)
    # Interpretation
    if res['reject_H0']:
        print("Reject H0: sample provides evidence that mean != {:.3f}".format(popmean))
    else:
        print("Fail to reject H0: no evidence that mean differs from {:.3f}".format(popmean))
```

```python
#Question 8: Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.
# clt_simulation.py
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns

def clt_demo(dist_rvs, sample_sizes=[1,5,10,30,50], n_sims=10000):
    results = {}
    for n in sample_sizes:
        means = [np.mean(dist_rvs(n)) for _ in range(n_sims)]
        results[n] = np.array(means)
    return results

if __name__ == "__main__":
    rng = np.random.default_rng(123)
    # Example: Exponential (skewed)
    def exp_rvs(n):
        return rng.exponential(scale=1.0, size=n)

    sample_sizes = [1, 5, 10, 30, 50]
    res = clt_demo(exp_rvs, sample_sizes=sample_sizes, n_sims=5000)

    plt.figure(figsize=(12,8))
    for i, n in enumerate(sample_sizes):
        plt.subplot(2, 3, i+1)
        sns.histplot(res[n], stat='density', kde=True)
        # overlay normal with same mean and var
        m = np.mean(res[n])
        s = np.std(res[n])
        x = np.linspace(min(res[n]), max(res[n]), 200)
        plt.plot(x, stats.norm.pdf(x, m, s), linestyle='--')
        plt.title(f'n={n}, mean‚âà{m:.3f}, sd‚âà{s:.3f}')
    plt.tight_layout()
    plt.show()

```

```python
#Question 9: Write a Python function to calculate and visualize the confidence interval for a sample mean.
# ci_visualize.py
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def mean_confidence_interval(sample, confidence=0.95):
    """
    Returns (mean, lower, upper) using t-distribution (unknown sigma).
    """
    x = np.array(sample)
    n = len(x)
    mean = np.mean(x)
    se = np.std(x, ddof=1) / np.sqrt(n)
    df = n - 1
    alpha = 1 - confidence
    t_crit = stats.t.ppf(1 - alpha/2, df)
    moe = t_crit * se
    return mean, mean - moe, mean + moe

def plot_ci(sample, confidence=0.95):
    mean, lo, hi = mean_confidence_interval(sample, confidence)
    plt.figure(figsize=(6,3))
    plt.errorbar(1, mean, yerr=[[mean-lo], [hi-mean]], fmt='o', capsize=8)
    plt.xlim(0,2)
    plt.xticks([])
    plt.ylabel('Sample mean')
    plt.title(f'{int(confidence*100)}% CI for mean: [{lo:.3f}, {hi:.3f}]')
    plt.show()

# Example
if __name__ == "__main__":
    import numpy as np
    sample = np.random.normal(loc=10, scale=2, size=30)
    mean, lo, hi = mean_confidence_interval(sample, 0.95)
    print(f"Mean = {mean:.3f}, 95% CI = [{lo:.3f}, {hi:.3f}]")
    plot_ci(sample, 0.95)
```

```python
#Question 10: Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.
# chi_square_gof.py
import numpy as np
import scipy.stats as stats

def chi_square_gof(observed_counts, expected_probs=None, expected_counts=None, ddof=0):
    """
    observed_counts: array-like observed counts per category
    expected_probs: if given, expected probabilities per category (sum to 1)
    expected_counts: if given, expected counts per category (sum should equal sum(observed))
    ddof: degrees of freedom adjustment (default 0). For estimated parameters, increase ddof.
    Returns chi2_stat, p_value, df
    """
    obs = np.array(observed_counts)
    n = obs.sum()

    if expected_counts is None:
        if expected_probs is None:
            raise ValueError("Either expected_probs or expected_counts must be provided.")
        expected_counts = np.array(expected_probs) * n

    expected_counts = np.asarray(expected_counts)
    if expected_counts.shape != obs.shape:
        raise ValueError("Expected and observed must have same shape.")

    chi2_stat = ((obs - expected_counts)**2 / expected_counts).sum()
    k = len(obs)
    df = k - 1 - ddof
    p_value = 1 - stats.chi2.cdf(chi2_stat, df)
    return chi2_stat, p_value, df

if __name__ == "__main__":
    # Example: observed counts of dice rolls (suppose 600 rolls)
    observed = np.array([95, 98, 102, 100, 103, 102])  # sums to 600
    expected_probs = np.ones(6) / 6  # fair die
    chi2, p, df = chi_square_gof(observed, expected_probs=expected_probs)
    print(f"Chi2 = {chi2:.3f}, df = {df}, p-value = {p:.4f}")
    if p < 0.05:
        print("Reject H0: observed distribution differs from expected.")
    else:
        print("Fail to reject H0: no evidence against expected distribution.")
```
